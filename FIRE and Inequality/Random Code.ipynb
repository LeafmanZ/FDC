```python
ERP_GDP_Industry = os.path.join(base_dir, 'ERP_GDP_INDUSTRY.ods')
```

```python
import pandas as pd
```

```python
# Data is GDP by Industry
# Scraped from Economic Report of the President using
# 2021 Table B-8 "Used years 1997-2019"
# 2006 Table B-12 "Used years 1974-1996"
gdpi_df = pd.read_excel(ERP_GDP_Industry)
```

```python
gdpi_df.head()
```

```python
index = []

for col in gdpi_df.columns:
    index.append(col)
    
index.remove('YEAR')
index
```

```python
gdpir_df = gdpi_df.copy()

for i in index:
    gdpir_df[i] = gdpir_df[i]/gdpi_df['GDP']
    
gdpir_df[index] = gdpir_df[index] * 100    
gdpir_df.head()
```

```python
removal_list = ['GDP', 'TOTAL PRIVATE INDUSTRIES', 
                'PRIVATE GOODS PRODUCING INDUSTRIES', 'PRIVATE SERVICES PRODUCING INDUSTRIES']

for i in removal_list:
    index.remove(i)

index
```

```python
gdpiri_df = gdpir_df[index]
gdpiri_df.head()
```

```python
gdpiri_df.sum(axis=1).head()
```

```python
CENSUS_SHR_Income = os.path.join(base_dir, 'CENSUS_SHR_INCOME.ods')
```

```python
# Data is Share of Aggregate Income Received by Each Fifth and Top 5 Percent of Households
# Scraped from US Census Bureau - Historical Income Tables: Households using
# Table H-2. Share of Aggregate Income Received by Each Fifth and Top 5 Percent of Households
sair_df = pd.read_excel(CENSUS_SHR_Income)
```

```python
sair_df.head()
```

```python
sair_df = sair_df[::-1].reset_index()
sair_df.head()
```

```python
if (len(sair_df.index) != len(gdpiri_df.index)):
    print('ERROR: YEARS ARE NOT ALIGNED!')
    print('Share of Aggregate Income Received (sair_df) length:', len(sair_df.index))
    print('GDP by Industry as a Ratio Industry only (gdpiri_df) length:', len(sair_df.index))
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
X = gdpiri_df['FINANCE, INSURANCE, REAL ESTATE, RENTAL, AND LEASING'].to_numpy()
y = sair_df['HIGHESTFIFTH'].to_numpy()

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, train_size = 0.7, 
                                                    test_size = 0.3, random_state = 100)
```

```python
# Importing Statsmodels.api library from Stamodel package
import statsmodels.api as sm

# Adding a constant to get an intercept
X_train_sm = sm.add_constant(X_train)
```

```python
# Fitting the resgression line using 'OLS'
lr = sm.OLS(y_train, X_train_sm).fit()

# Printing the parameters
lr.params
```

```python
lr.summary()
```

```python
plt.scatter(X_train, y_train)
plt.plot(X_train, lr.params[0] + lr.params[1]*X_train, 'r')
plt.title('Training plot')
plt.xlabel('Percent FIRE of GDP')
plt.ylabel('Highest Fifth Share of Aggregate Income')
plt.show()
```

```python
y_train_pred = lr.predict(X_train_sm)
res = (y_train - y_train_pred)
```

```python
# Adding a constant to X_test
X_test_sm = sm.add_constant(X_test)

# Predicting the y values corresponding to X_test_sm
y_test_pred = lr.predict(X_test_sm)

# Printing the first 15 predicted values
y_test_pred
```

```python
# Importing r2_square
from sklearn.metrics import r2_score

# Checking the R-squared value
r_squared = r2_score(y_test, y_test_pred)
r_squared
```

```python
# Visualize the line on the test set
plt.scatter(X_test, y_test)
plt.plot(X_test, y_test_pred, 'r')
plt.title('Testing plot')
plt.xlabel('Percent FIRE of GDP')
plt.ylabel('Highest Fifth Share of Aggregate Income')
plt.show()
```

```python
plt.scatter(X, y)
plt.plot(X, lr.params[0] + lr.params[1]*X, 'r')
plt.title('Full plot')
plt.xlabel('Percent FIRE of GDP')
plt.ylabel('Highest Fifth Share of Aggregate Income')
plt.show()
```
